# Windowing

```python
#Working on the groundTruth dataset
ground_truth=pd.read_csv("groundTruth.csv",sep=",") #reading the ground truth csv file
#ground_truth.head()
event_size=len(ground_truth.axes[0])    #No of events described in ground truth
start=ground_truth[' inicio'] *(10**9)  #starting time of every event
end=ground_truth[' fim']*(10**9)        #end time for each event label
                                        #Conversion from seconds to nanoseconds
samples=start.shape[0]
```

```python
#Setting the size of window..
window_size=10
```

## Following codes is to built a Nan dataset with columns in the order as shown
## Mean(x)(0,1,2,..size-1),Median(x)...SD(x)(...),,,Tendency(x)(...),..Mean(y)........,,,........Mean(z)....,,,,,...,,,..,Tendency(z)(1,2...size-1)

```python
dataset = pd.DataFrame(index=range(0,samples*window_size))
test=pd.DataFrame(np.nan, index=range(0,160),columns=['Labels'])
```

```python
for axis in ['x','y','z']:
    for op in ['Mean_','Median_','SD_','Tendency_']:
        for i in range(0,window_size):
            if op!='Tendency_':
                dataset[str(op)+str(axis)+str(i)]=test
            elif i>0:
                dataset[str(op)+str(axis)+str(i)]=test
dataset['Labels']=test
#dataset['Labels']=dataset['Labels'].astype(str)
```

```python
dataset
```

```python
def attribute(w_point,index,axis):
    list=[]
    for i in range(0,index+1):
        list.append(raw_data[str(axis)][w_point-i-1])
    #print(np.mean(list))    
    return np.mean(list),np.median(list),np.std(list);   
```

```python
def Windowing(j,w_point,label):
    dataset['Labels'][j]=label
    for axis in ['x','y','z']:
        for op in ['Mean_','Median_','SD_']:
            i=0
            #tup=()
            while i<window_size:
                #print(i)
                #print(dataset['Mean_'+str(axis)+str(i)][j])
                #print(type(dataset['Mean_'+str(axis)+str(i)][j]))
                #tup=attribute(w_point,i,axis)
                #print(type(tup[0]))
                (dataset['Mean_'+str(axis)+str(i)][j],dataset['Median_'+str(axis)+str(i)][j],dataset['SD_'+str(axis)+str(i)][j])=attribute(w_point,i,axis)
                #dataset[str(op)+str(axis)+str(window_size-i)][j]=op(w_point,i)
                #print(new_frame[str(op)+str(axis)][(int(w_point))+window_size-i])
                i+=1
            if op == 'Mean_':
                for t in range(1,window_size):
                    dataset['Tendency_'+str(axis)+str(t)][j]=dataset['Mean_'+str(axis)+str(t)][j]/dataset['Mean_'+str(axis)+str(0)][j]
```

```python
j=0
for i in range(0,len(start)):
    s=start[i]
    w_point=int(s)+1000000000
    label=ground_truth['evento'][i]
    print(label)
    print('w_point='+str(w_point))
    while (s/1000000000)>=((w_point/1000000000)-window_size):
        try:
            Windowing(j,int(w_point/1000000000),label)
            #Windowing(j,int(w_point/1000000000))
        except:
            w_point=window_size*10**9
            #Windowing(j,int(w_point/1000000000))
            Windowing(j,int(w_point/1000000000),label)
        j+=1
        w_point+=1000000000
```

```python
dataset=dataset.dropna(axis=0,how='any')
```
