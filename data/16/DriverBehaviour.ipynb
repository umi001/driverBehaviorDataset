# FRAMING
 Framing the entire dataset to one second frames from multiple nanosecond frames

```python
#Created a new dataframe to add all the new frame attributes.This will form our new dataset.
new_frame = pd.DataFrame(np.nan, index=range(0,1269),
                         columns=['time','Mean_x','Mean_y','Mean_z',
                                  'Median_x','Median_y','Median_z','SD_x','SD_y','SD_z']) 
new_frame
```

```python
def framing(a):
    if a=='x':
        col=1
    elif a=='y':
        col=2
    elif a=='z':
        col=3    
        
    t=1
    j=0
    for i in range(0,1269):
        list=[]
        while (raw_data['uptimeNanos'][j])<=(t*10**9):
            list.append(raw_data[a][j])
            j+=1
        new_frame.iloc[:,col][t-1]=np.mean(list)
        new_frame.iloc[:,col+3][t-1]=np.median(list)
        new_frame.iloc[:,col+6][t-1]=np.std(list)
        new_frame['time'][t-1]=t
        t+=1
        
```

```python
framing('x')
framing('y')
framing('z')
```
